# Method For Numpy Array Creation

```python
zero=np.zeros((3,4))  # 3 rows and 4 columns
```

```python
zero # A 3x4 array consist of all zeros
```

```python
custom_array=np.full((2,3),10)  # 2 rows and 3 columns filled with ones,10 by default
```

```python
custom_array
```

```python
custom_array.shape # Shape of the array
```

```python
one_array=np.ones((2,3))  # 2 rows and 3 columns filled with ones
```

```python
one_array
```

```python
array=np.arange(10)  # Array with values ranging from 0 to 9
```

```python
array
```

```python
lin_array=np.linspace(0,5,10)  # Array with 10 values ranging from 0 to 5
```

```python
lin_array
```

```python
identy_array=np.eye(4)  # 4x4 Identity matrix
```

```python
identy_array
```

```python
empty_array=np.empty((2,3))  # 2 rows and 3 columns uninitialized
```

```python
empty_array
```

```python
id_array=np.identity(5)  # 5x5 Identity matrix
```

```python
id_array
```
