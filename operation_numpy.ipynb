```python
import numpy as np
```

```python
array1=np.array([1,2,3])
array2=np.array([4,5,6])
```

```python
# addition
array_sum=array1+array2
array_sum
```

```python
# Subtraction
array_sub=array2-array1
array_sub
```

```python
# Multiplication
array_mul=array1*array2
array_mul
```

```python
# Division
array_div=array2/array1
array_div
```

```python
# // operation
array_floor_div=array2//array1
array_floor_div
```

```python
# Square root
array_sqrt=np.sqrt(array1)
array_sqrt
```

```python
# sum
array_total_sum=np.sum(array1)
array_total_sum
```

```python
# min 
array_min=np.min(array1)
array_min
```

```python
# max
array_max=np.max(array1)
array_max
```

```python
# find element specific condition
array_specific_Condition_element=array1.where(array1>2)
array_specific_Condition_element
```

```python
# count non zero in array
array_count_non_zero=np.count_nonzero(array1)
array_count_non_zero
```

```python
# Merge
array_merged=np.concatenate((array1,array2))
array_merged
```

```python
# merge row wise
array_2d_1=np.array([[1,2,3],[4,5,6]])
array_2d_2=np.array([[7,8,9],[10,11,12]])
```

```python
mergred_row_wise=np.vstack((array_2d_1,array_2d_2))
mergred_row_wise
```

```python
merged_colums_wise=np.hstack((array_2d_1,array_2d_2))
merged_colums_wise
```

```python
# reverse array
array_reversed=array1.flip()
array_reversed
```
